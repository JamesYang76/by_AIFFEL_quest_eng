## 데이터 다운로드
아래 링크에서 korean-english-park.train.tar.gz 를 다운로드받아 한영 병렬 데이터를 확보합니다.

- [jungyeul/korean-parallel-corpora](https://github.com/jungyeul/korean-parallel-corpora/tree/master/korean-english-news-v1)

````python
import requests
import tarfile

save_dir = os.path.expanduser("work/s2s_translation/datasets")
filename = "korean-english-park.train.tar.gz"
url = "https://raw.githubusercontent.com/jungyeul/korean-parallel-corpora/master/korean-english-news-v1/korean-english-park.train.tar.gz"

# 전체 파일 경로 만들기 (운영체제에 맞게 경로 결합)
file_path = os.path.join(save_dir, filename)

# 파일일 없으면 파일 다운로드
if not os.path.exists(file_path):

    # 디렉토리(폴더)가 없으면 생성
    # exist_ok=True: 폴더가 이미 있어도 에러를 내지 않음
    os.makedirs(save_dir, exist_ok=True)

    print(f"'{save_dir}' 경로에 다운로드를 시작합니다...")

    # 다운로드 실행
    response = requests.get(url, stream=True)

    if response.status_code == 200:
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=1024):
                f.write(chunk)
        print(f"다운로드 완료! 저장 위치: {file_path}")
    else:
        print(f"다운로드 실패. 상태 코드: {response.status_code}")


data_folder = os.path.join(save_dir, "kor-eng")
en_path_to_file = os.path.join(data_folder, "korean-english-park.train.en")
ko_path_to_file = os.path.join(data_folder, "korean-english-park.train.ko")

if not os.path.exists(en_path_to_file):
    try:
        print("압축 해제 중...")

        with tarfile.open(file_path, "r:gz") as tar:  # 'r:gz' 모드는 gzip으로 압축된 tar 파일을 읽겠다는 의미입니다.
            tar.extractall(path=data_folder)

        print("압축 해제 완료!")

    except Exception as e:
        print(f"오류 발생: {e}")

print("생성된 파일: ", os.listdir(data_folder))
````

````python
# 줄바꿈 기준으로 5줄 출력 함수

def print_5line(contents):
    lines = contents.splitlines() # 줄바꿈을 기준으로 나누어 리스트 생성
    for line in lines[:5]:  # 처음 5개 요소(5줄)를 출력
        print(line)
````

````python
def preprocess_sentence(sentence):
    sentence = sentence.lower().strip()

    sentence = re.sub(r"([?.!,])", r" \1 ", sentence)
    sentence = re.sub(r'[" "]+', " ", sentence)
    sentence = re.sub(r"[^가-힣a-zA-Z0-9?.!,]+", " ", sentence)  # 한글, 숫자 추가

    sentence = sentence.strip()

    return sentence

print("슝~")
````

````python
# 상위 3만 개만 사용 -> 6만개

df = df_sample[:60000]
# df = df_sample.copy()

df["Korean"] = df["Korean"].apply(preprocess_sentence)
df["English"] = df["English"].apply(lambda x: preprocess_sentence(x))

df.head()
````

## 데이터 전처리: 토큰화

````python
# 토큰 길이 40 -> 60 -> 80 이하인 데이터만 선별 (모든 데이터를 사용할 경우 학습에 굉장히 오랜 시간이 걸립니다.)
mask = (
    df["Korean"].str.split(" ").str.len() <= 80   # <= 40  # 한국어 토큰 길이
) & (
    df["English"].str.split(" ").str.len() <= 80  # <= 40  # 영어 토큰 길이
)

# 필터링된 데이터프레임
filtered_df = df[mask].copy()
````

````python
MECAB_TOKEN = 1
OKT_TOKEN = 2
NO_PRE_TOKEN = 3

Mecab_Okt_option = OKT_TOKEN

# English corpus 저장
filtered_df["English"].to_csv("p2_eng_corpus.txt", index=False, header=False, sep="\n", encoding="utf-8")

# Korean corpus 저장
if Mecab_Okt_option == MECAB_TOKEN:  # Mecab 사용 시
    mecab = Mecab()

    # filtered_df["Korean"]에 한글 문장 리스트가 있다고 가정
    with open('korean_mecab.txt', 'w', encoding='utf-8') as f:
        for kor_sentence in filtered_df["Korean"]:

            # Mecab으로 형태소 분리 (예: "아버지가 방에" -> ['아버지', '가', '방', '에'])
            morphs = mecab.morphs(kor_sentence)

            # 공백으로 합쳐서 한 줄에 저장 (예: "아버지 가 방 에")
            # SentencePiece는 텍스트 파일을 입력으로 받기 때문
            f.write(' '.join(morphs) + '\n')

    print("Mecab 전처리 완료! 'korean_mecab.txt' 생성됨.")
    print(f"총 {len(filtered_df)} 문장 처리됨.")
    print("파일 저장 완료: p2_eng_corpus.txt, korean_mecab.txt")

elif Mecab_Okt_option == OKT_TOKEN:  # Okt 사용 시
   # 1. Okt 객체 초기화
    okt = Okt()

    # 2. 전처리 및 파일 저장
    with open('korean_okt.txt', 'w', encoding='utf-8') as f:
        for kor_sentence in filtered_df["Korean"]:

            # Okt로 형태소 분석 (예: "아버지가" -> ['아버지', '가'])
            # stem=True 옵션을 주면 '가고있다' -> '가다'처럼 원형 복원이 가능합니다 (선택사항)
            morphs = okt.morphs(kor_sentence)

            # 형태소들을 공백으로 구분된 문자열로 변환하여 파일에 쓰기
            f.write(' '.join(morphs) + '\n')

    print("Okt 전처리 완료! 'korean_okt.txt' 생성됨.")
    print(f"총 {len(filtered_df)} 문장 처리됨.")
    print("파일 저장 완료: p2_eng_corpus.txt, korean_okt.txt")

else:    # Mecab이나 Okt 미사용
    filtered_df["Korean"].to_csv("p2_kor_corpus.txt", index=False, header=False, sep="\n", encoding="utf-8")
    filtered_df["English"].to_csv("p2_eng_corpus.txt", index=False, header=False, sep="\n", encoding="utf-8")

    print("파일 저장 완료: p2_eng_corpus.txt, p2_kor_corpus.txt")
````

### SentencePiece를 활용해 Tokenizer를 생성하고 테이터를 변환

#### 1. SentencePiece 주요 역할 (핵심 기능)
- 인코딩 (Encoding): 문장을 토큰(Token) 단위로 쪼개고, 각 토큰을 고유한 숫자(Index)로 변환합니다.
    - 예: "아버지가 방에" → ['아버지', '가', '방', '에'] → [105, 22, 403, 9]
- 디코딩 (Decoding): 모델이 내놓은 숫자 리스트를 다시 원래의 문장으로 합칩니다.
    - 예: [105, 22, 403, 9] → "아버지가 방에"
- 어휘 사전 관리 (Vocabulary Management): 학습 데이터를 통해 가장 자주 쓰이는 단어/서브워드 조각들을 모아 vocab 파일을 만들고 관리합니다.

#### 2. SentencePiece 토크나이저만의 차별점
SentencePiece는 일반적인 형태소 분석기(Mecab, Okt 등) 기반 토크나이저와 작동 방식이 조금 다릅니다.

① 공백도 문자로 취급 (언더바 _ 활용)
기존 방식은 공백을 기준으로 단어를 나누고 공백을 버리지만, SentencePiece는 공백을 _(U+2581)라는 특수 문자로 치환하여 보존합니다. 덕분에 디코딩할 때 원래 문장의 띄어쓰기를 완벽하게 복원할 수 있습니다.

② 서브워드(Subword) 분리 (BPE/Unigram)
단어를 더 작은 단위로 쪼개어 '미등록 단어(OOV, Out-Of-Vocabulary)' 문제를 해결합니다.

예를 들어 "신조어"라는 단어가 사전에 없어도, "신", "조", "어"라는 글자 조각들이 사전에 있다면 이를 조합해서 인식합니다.

③ 언어 독립성 (Language Independent)
Mecab처럼 특정 언어(한국어)의 문법 규칙을 미리 알 필요가 없습니다. 단순히 데이터의 통계적 빈도를 기반으로 쪼개기 때문에 한국어, 영어, 일본어 심지어 이모지까지 동일한 로직으로 처리합니다.

#### 3. 작업 흐름도
SentencePiece 작업은 크게 **학습(Train)**과 추론(Inference) 단계로 나뉩니다.
- Train: 텍스트 뭉치(Corpus)를 읽어 빈도수가 높은 조각들을 모아 .model과 .vocab 파일을 만듭니다.
- Inference (Tokenizer 역할): 생성된 .model 파일을 불러와 실시간으로 문장을 쪼개고 숫자로 바꿉니다.

#### 4. Mecab이나 Okt로 미리 형태소를 나눈 뒤 SentencePiece를 돌리는 이유
**"한국어의 특성을 미리 반영해 주어 SentencePiece가 더 효율적인 조각을 찾도록 돕기 위해서"**입니다. (이를 Pre-tokenization이라고 합니다.)

````python
vocab_size = 20000  # 9914 # 오류 메시지에 따라 vocab_size를 9914 이하로 수정 (korean_mecab.txt 코퍼스에서는 최대 9914개의 고유한 단어만 추출)
pad_id = 0
bos_id = 1
eos_id = 2
unk_id = 3

if Mecab_Okt_option == MECAB_TOKEN:  # MeCab 사용 시
    input_file = "korean_mecab.txt"
    model_prefix = "p3_encoder_spm"

elif Mecab_Okt_option == OKT_TOKEN:  # Okt 사용 시
    input_file = "korean_okt.txt"
    model_prefix = "p4_encoder_spm"

else:  # 그 외
    input_file = "p2_kor_corpus.txt"
    model_prefix = "p2_encoder_spm"

spm.SentencePieceTrainer.train(
    input = input_file,
    model_prefix = model_prefix,
    vocab_size = vocab_size,
    pad_id = pad_id,
    bos_id = bos_id,
    eos_id = eos_id,
    unk_id = unk_id
)

spm.SentencePieceTrainer.train(
    input = "p2_eng_corpus.txt",
    model_prefix = "p2_decoder_spm",
    vocab_size = vocab_size,
    pad_id = pad_id,
    bos_id = bos_id,
    eos_id = eos_id,
    unk_id = unk_id
)
````

### `Encoder`와 `SentencePieceTrainer`의 관계

`Encoder`와 `SentencePieceTrainer`는 직접적으로 코드로 연결되어 있지는 않지만, 시퀀스-투-시퀀스 모델의 전체 워크플로우에서 서로 밀접하게 연결된 역할을 합니다.

1.  **`SentencePieceTrainer`의 역할 (전처리 단계)**:
    *   `SentencePieceTrainer.train()` 함수는 원본 텍스트 데이터(코퍼스, 예: `korean_okt.txt`)를 학습하여 **SentencePiece 모델 파일(`.model`)**과 **어휘 사전 파일(`.vocab`)**을 생성합니다.
    *   이 과정에서 `input_dim` (`vocab_size`) 파라미터로 지정된 크기만큼의 고유한 토큰과 그에 해당하는 ID가 결정됩니다. 예를 들어, 코드에서 `vocab_size = 20000`으로 설정되어 있습니다.
    *   생성된 `.model` 파일을 `spm.SentencePieceProcessor()` 객체(예: `encoder_tokenizer`)에 로드하여 실제 토큰화 작업을 수행할 수 있는 토크나이저를 만듭니다. 이 토크나이저가 한국어 문장을 숫자의 시퀀스(토큰 ID)로 변환하는 역할을 합니다.

2.  **`Encoder`의 역할 (모델 아키텍처의 일부)**:
    *   `Encoder` 클래스는 모델의 입력 부분을 담당합니다. 특히, `self.embedding = nn.Embedding(input_dim, emb_dim)` 이 부분이 `SentencePieceTrainer`가 생성한 결과물과 연결됩니다.
    *   여기서 `input_dim`은 `encoder_tokenizer`가 학습한 **전체 어휘 사전의 크기**를 의미합니다. `nn.Embedding` 레이어는 이 어휘 사전 크기만큼의 고유한 토큰 ID를 입력으로 받을 준비를 합니다.
    *   `Encoder`의 `forward` 메서드로 들어오는 `src` (소스 시퀀스)는 `encoder_tokenizer`에 의해 이미 숫자의 시퀀스(토큰 ID)로 변환된 상태입니다.
    *   `self.embedding(src)`는 이러한 토큰 ID들을 받아서, 각각의 ID에 해당하는 **밀집 임베딩 벡터**로 변환합니다. 이 임베딩 벡터들은 단어의 의미 정보를 포함하며, 이후 LSTM 레이어의 입력으로 사용됩니다.

**요약하자면,**
`SentencePieceTrainer`는 텍스트를 모델이 이해할 수 있는 숫자(토큰 ID)로 변환하기 위한 '규칙'과 '사전'을 만드는 도구입니다. 이렇게 만들어진 토큰 ID들은 `Encoder`의 `nn.Embedding` 레이어의 입력으로 들어가, 학습 가능한 '의미 벡터'로 변환되어 모델 학습에 사용됩니다.

따라서 `SentencePieceTrainer`가 없이는 `Encoder`의 `nn.Embedding` 레이어가 어떤 토큰을 몇 개의 숫자로 표현해야 할지 알 수 없으며, 모델은 텍스트 데이터를 처리할 수 없게 됩니다.

````python
model_name = model_prefix + ".model"
encoder_tokenizer = spm.SentencePieceProcessor()
encoder_tokenizer.load(model_name)

decoder_tokenizer = spm.SentencePieceProcessor()
decoder_tokenizer.load("p2_decoder_spm.model")
````

### 데이터셋 생성
START_TOKEN, END_TOKEN 추가 + Padding

````python
class TranslationDataset(Dataset):
    def __init__(self, data, encoder_tokenizer, decoder_tokenizer, max_len):
        self.data = data
        self.encoder_tokenizer = encoder_tokenizer
        self.decoder_tokenizer = decoder_tokenizer
        self.max_len = max_len
        self.pad_id = 0
        self.bos_id = 1
        self.eos_id = 2

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        src_text = self.data.iloc[idx]['Korean']
        trg_text = self.data.iloc[idx]['English']

        src_ids = self.encoder_tokenizer.encode(src_text)
        trg_ids = self.decoder_tokenizer.encode(trg_text)

        src_ids = src_ids[:self.max_len]

        # Decoder의 입력에는 START_TOKEN과 END_TOKEN을 추가해줍니다. 단, 최대 길이 제한을 적용시킵니다.
        trg_input = [self.bos_id] + trg_ids[:self.max_len - 2] + [self.eos_id]
        trg_label = trg_ids[:self.max_len - 1] + [self.eos_id]

        # 길이가 짧은 경우 PAD_TOKEN을 추가해줍니다.
        src_ids = src_ids + [self.pad_id] * (self.max_len - len(src_ids))
        trg_input = trg_input + [self.pad_id] * (self.max_len - len(trg_input))
        trg_label = trg_label + [self.pad_id] * (self.max_len - len(trg_label))

        return torch.tensor(src_ids), torch.tensor(trg_input), torch.tensor(trg_label)
````

## 3. 모델 설계
### Bahdanau Attention 설계

````python
class BahdanauAttention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()

        self.W1 = nn.Linear(hidden_dim, hidden_dim)
        self.W2 = nn.Linear(hidden_dim, hidden_dim)
        self.v = nn.Linear(hidden_dim, 1, bias=False)

    def forward(self, hidden, encoder_outputs):
        # hidden: 타겟 시퀀스의 다음 단어를 생성하기 위해 현재 디코더가 '어떤 정보'를 가지고 있는지를 나타내며,
        #         이 정보를 바탕으로 인코더 출력 중 어느 부분에 '집중'할지를 결정하는 데 사용
        # encoder_outputs: 인코더의 모든 은닉 상태 출력

        # hidden: (1, batch_size, hidden_dim) 또는 (batch_size, hidden_dim)으로 올 수 있음
        # LSTM hidden state는 (num_layers, batch_size, hidden_dim) 형태이므로, 첫 번째 차원(레이어)을 제거합니다.
        if hidden.dim() == 3:
            hidden = hidden.squeeze(0) # (batch_size, hidden_dim)

        # encoder_outputs: (batch_size, src_len, hidden_dim) (Encoder에서 batch_first=True로 오기 때문에)

        # src_len을 encoder_outputs의 두 번째 차원(원본 src_len)에서 가져옵니다.
        src_len = encoder_outputs.shape[1] # <--- 수정: shape[0] 대신 shape[1]

        # hidden을 (batch_size, src_len, hidden_dim) 형태로 확장합니다.
        hidden_for_attention = hidden.unsqueeze(1).repeat(1, src_len, 1) # (batch_size, src_len, hidden_dim)

        energy = torch.tanh(self.W1(encoder_outputs) + self.W2(hidden_for_attention))  # (batch_size, src_len, hidden_dim)
        attention = self.v(energy).squeeze(2)  # (batch_size, src_len)

        return nn.functional.softmax(attention, dim=1)  # (batch_size, src_len)
        # 어텐션 스코어를 계산
        # 1. hidden을 encoder_outputs의 src_len에 맞춰 확장합니다. (hidden_for_attention)
        # 2. W1(encoder_outputs)와 W2(hidden_for_attention)를 통해 각각을 변환한 후 더하고 tanh를 적용하여 '에너지'(energy)를 계산합니다.
        # 3. 이 energy에 v 선형 레이어를 적용하여 각 소스 시퀀스 위치에 대한 스칼라 어텐션 점수(attention)를 얻습니다.
        # 4. 마지막으로 softmax를 적용하여 이 점수들을 정규화된 확률 분포(어텐션 가중치)로 변환합니다.
        # 이 가중치들이 바로 디코더가 현재 출력 단어를 생성하기 위해 소스 시퀀스의 어느 부분에 더 집중해야 하는지를 알려주는 정보가 됩니다.
````

### BahdanauAttention 클래스의 파라미터 설명

`BahdanauAttention` 클래스에서 사용되는 `batch_size`, `src_len`, `hidden_dim`은 신경망, 특히 시퀀스-투-시퀀스(Seq2Seq) 모델의 어텐션 메커니즘에서 중요한 개념들입니다.

1.  **`batch_size`**: 모델이 한 번에 처리하는 시퀀스의 개수를 나타냅니다.
    *   딥러닝 모델은 일반적으로 효율성을 위해 데이터를 여러 개 묶어 배치(batch) 단위로 처리합니다.
    *   예를 들어, `batch_size`가 64라면, 64개의 문장 쌍(예: 64개의 한국어 문장과 그에 대응하는 64개의 영어 문장)이 동시에 모델에 입력되어 처리됩니다.
    *   `BahdanauAttention` 코드에서 `encoder_outputs`의 첫 번째 차원이 `batch_size`에 해당합니다: `encoder_outputs: (batch_size, src_len, hidden_dim)`.

2.  **`src_len` (Source Sequence Length)**: 원본 입력 시퀀스(여기서는 한국어 문장)의 길이를 나타냅니다. 즉, 원본 문장을 토큰화했을 때의 토큰 개수입니다.
    *   어텐션 메커니즘은 디코더가 타겟 시퀀스의 다음 단어를 예측할 때, 원본 시퀀스의 어느 부분에 더 집중해야 하는지를 결정합니다.
    *   `src_len`은 인코더가 출력한 각 토큰에 대한 정보를 나타내는 벡터의 개수와 같습니다.
    *   `BahdanauAttention` 코드에서 `encoder_outputs`의 두 번째 차원이 `src_len`에 해당합니다: `encoder_outputs: (batch_size, src_len, hidden_dim)`.

3.  **`hidden_dim` (Hidden Dimension)**: RNN(여기서는 LSTM)의 은닉 상태(hidden state) 벡터의 차원(크기)을 나타냅니다.
    *   이 값은 RNN 셀이 이전 정보를 얼마나 많은 숫자로 압축하여 저장하는지를 결정합니다. 즉, 각 타임스텝의 출력 및 은닉 상태의 '정보량'을 나타내는 차원입니다.
    *   `hidden_dim`이 클수록 모델이 더 복잡한 패턴을 학습할 수 있지만, 계산 비용이 증가하고 과적합의 위험이 있습니다.
    *   `BahdanauAttention` 코드에서 `hidden` (디코더의 현재 은닉 상태)과 `encoder_outputs`의 마지막 차원이 `hidden_dim`에 해당합니다: `hidden: (batch_size, hidden_dim)`, `encoder_outputs: (batch_size, src_len, hidden_dim)`.

### Encoder, Decoder 설계

````python
class Encoder(nn.Module):
    def __init__(self, input_dim, emb_dim, hidden_dim, num_layers):
        super().__init__()

        self.embedding = nn.Embedding(input_dim, emb_dim)
        # batch_first=False로 변경 (또는 embedded.permute(1, 0, 2)를 forward에서 추가)
        self.rnn = nn.LSTM(emb_dim, hidden_dim, num_layers, batch_first=True)  # =False)

    def forward(self, src):
        # src : (src_len, batch_size)  # src_len: 원본(source) 언어의 토큰수(문장의 길이),  batch_size: 문장의 개수
        embedded = self.embedding(src)  # embedded : (src_len, batch_size, emb_dim)

        # 만약 batch_first=True를 유지하고 싶다면 아래 라인의 주석을 해제
        embedded = embedded.permute(1, 0, 2) # (batch_size, src_len, emb_dim)

        outputs, (h_0, c_0) = self.rnn(embedded)  # (batch_size, src_len, hidden_dim) (batch_first=True이고 embedded를 permute한 경우)
                                                  # 또는 outputs : (src_len, batch_size, hidden_dim) (batch_first=False일 경우)
        # h_0, c_0 : (1, batch_size, hidden_dim)  # num_layers * num_directions, batch, hidden_dim

        return outputs, h_0, c_0
````

### 인코더 설명

#### 1. Encoder(nn.Module)의 Embedding 설명

`Encoder` 클래스 내에서 `nn.Embedding`은 입력 시퀀스(여기서는 한국어 문장)의 각 토큰을 밀집 벡터(dense vector)로 변환하는 역할을 합니다.

코드에서 다음과 같이 정의됩니다:

```python
class Encoder(nn.Module):
    def __init__(self, input_dim, emb_dim, hidden_dim, num_layers):
        super().__init__()
        self.embedding = nn.Embedding(input_dim, emb_dim)
        # ... (나머지 코드)
```

각 파라미터의 의미는 다음과 같습니다:

*   **`self.embedding = nn.Embedding(input_dim, emb_dim)`**:
    *   `nn.Embedding`: PyTorch에서 제공하는 임베딩 레이어로, 정수 인덱스로 표현된 토큰들을 고차원의 연속적인 벡터 공간으로 매핑합니다.
    *   `input_dim` (또는 `vocab_size`): 인코더가 처리할 수 있는 전체 단어(토큰)의 개수, 즉 어휘 사전(vocabulary)의 크기입니다. 예를 들어, `encoder_tokenizer`의 어휘 사전 크기(`len(encoder_tokenizer)`)가 여기에 해당합니다.
    *   `emb_dim`: 각 단어(토큰)가 매핑될 임베딩 벡터의 차원(크기)입니다. 이 값은 각 단어의 의미적 특징을 얼마나 풍부하게 표현할지를 결정합니다. 예를 들어, `emb_dim = 256`은 각 토큰이 256차원 벡터로 표현된다는 의미입니다.

**역할과 중요성:**

1.  **희소 표현(One-Hot Encoding)의 한계 극복:** 자연어 처리에서 단어는 종종 원-핫 인코딩(one-hot encoding)과 같은 희소 표현으로 나타낼 수 있습니다. 하지만 이는 단어 간의 유사성을 표현하기 어렵고, 고차원이며, 벡터의 대부분이 0이어서 비효율적입니다.
2.  **의미적 정보 학습:** `nn.Embedding` 레이어는 각 단어에 대한 고유한 벡터를 학습합니다. 이 벡터들은 학습 과정에서 단어의 의미적, 문맥적 유사성을 반영하게 되어, 유사한 단어들은 벡터 공간에서 가까운 위치에 놓이게 됩니다.
3.  **RNN/LSTM 입력 준비:** 인코더의 `forward` 메서드에서 입력 `src` (토큰 ID들의 시퀀스)는 먼저 이 `embedding` 레이어를 통과하여 `embedded` (임베딩 벡터 시퀀스)로 변환됩니다. 이렇게 변환된 `embedded` 벡터들이 이후 `LSTM` 레이어의 입력으로 사용되어 문장의 정보를 추출하게 됩니다.

요약하자면, `Encoder`의 `nn.Embedding` 레이어는 한국어 입력 문장의 각 단어를 숫자의 나열(토큰 ID)에서 모델이 이해하고 처리할 수 있는 의미 있는 연속적인 벡터 표현으로 바꿔주는 핵심적인 초기 처리 단계입니다.

#### 2. `nn.LSTM`의 역할 설명

`nn.LSTM`은 PyTorch에서 제공하는 **Long Short-Term Memory(LSTM)** 네트워크 모듈입니다. LSTM은 순환 신경망(RNN)의 한 종류로, 특히 긴 시퀀스의 정보를 학습하고 기억하는 데 탁월한 성능을 보입니다.

`Encoder` 클래스 내에서 `self.rnn = nn.LSTM(emb_dim, hidden_dim, num_layers, batch_first=True)`와 같이 사용됩니다.

각 파라미터의 의미는 다음과 같습니다:

*   **`emb_dim` (Input Size)**: LSTM 레이어에 입력되는 각 시퀀스 요소(토큰)의 특징 차원입니다. `Encoder`에서는 `nn.Embedding` 레이어를 통해 변환된 임베딩 벡터의 차원(`emb_dim`)이 됩니다.
*   **`hidden_dim` (Hidden Size)**: LSTM 셀의 은닉 상태(hidden state) 및 셀 상태(cell state) 벡터의 차원입니다. 이 값이 클수록 모델이 더 복잡한 패턴을 학습할 수 있습니다.
*   **`num_layers`**: 스택될 LSTM 레이어의 개수입니다. 여러 개의 LSTM 레이어를 쌓으면 모델의 표현력을 높일 수 있습니다.
*   **`batch_first=True`**: 입력 및 출력 텐서의 첫 번째 차원이 배치 크기(batch size)임을 의미합니다. (즉, `(batch_size, sequence_length, features)` 형태).

**`Encoder` 내에서의 역할과 중요성:**

1.  **시퀀스 정보 학습**: `nn.LSTM`은 `nn.Embedding` 레이어를 거쳐 생성된 한국어 문장의 임베딩 벡터 시퀀스(`embedded`)를 입력으로 받습니다. 이 시퀀스를 시간 단계별로 순차적으로 처리하면서 문장의 문법적, 의미적 정보를 추출하고 학습합니다.
2.  **장기 의존성(Long-Term Dependencies) 처리**: 일반적인 RNN은 긴 시퀀스에서 앞부분의 정보가 뒷부분으로 갈수록 희미해지는 문제(vanishing gradient problem)를 겪지만, LSTM은 '게이트(gate)' 메커니즘(입력 게이트, 망각 게이트, 출력 게이트)을 사용하여 중요한 정보를 장기간 기억하고 불필요한 정보를 잊어버리면서 이러한 문제를 효과적으로 해결합니다.
3.  **문맥 벡터(Context Vector) 생성**: `Encoder`의 `nn.LSTM`은 입력 문장 전체를 처리한 후, 해당 문장의 종합적인 의미를 압축한 은닉 상태(hidden state)와 셀 상태(cell state)를 출력합니다. 이 은닉 상태와 셀 상태는 일반적으로 '문맥 벡터'로 사용되어 `Decoder`로 전달되며, `Decoder`가 번역 문장을 생성하는 데 필요한 정보를 제공합니다.

요약하자면, `Encoder`의 `nn.LSTM`은 한국어 원본 문장의 단어 임베딩 시퀀스를 입력받아, 그 문장의 복잡한 의미와 문맥을 이해하고 이를 고정된 크기의 벡터(은닉 상태)로 인코딩하는 핵심적인 역할을 수행합니다. 이 인코딩된 정보가 `Decoder`의 번역 과정에서 활용됩니다.

#### 3. `forward(self, src)` 메서드 설명

`forward` 메서드는 PyTorch의 `nn.Module`을 상속받는 모든 신경망 모델의 핵심입니다. 이 메서드는 모델이 입력 데이터를 받아 최종 출력을 생성하는 계산 그래프를 정의합니다.

1.  **`forward`는 언제 실행되는가?**
    *   `forward` 메서드는 모델 객체를 함수처럼 호출할 때 **자동으로** 실행됩니다.
    *   예를 들어, `encoder = Encoder(...)`와 같이 `Encoder` 클래스의 인스턴스를 생성한 후, `outputs, h_0, c_0 = encoder(src)`와 같이 `encoder` 객체를 호출하면 내부적으로 `encoder.forward(src)` 메서드가 실행됩니다.
    *   이것은 PyTorch의 `nn.Module` 클래스가 `__call__` 메서드를 구현하여 `forward`를 호출하도록 되어 있기 때문입니다. (`model(input)` 호출 → `model.__call__(input)` 실행 → `model.forward(input)` 실행).

2.  **`src`에는 무엇이 전달되는가?**
    *   `src`는 `Encoder` 모델의 **입력 데이터**입니다. 시퀀스-투-시퀀스 모델의 인코더에서는 보통 **원본(source) 언어의 시퀀스**를 나타냅니다.
    *   코드 상으로는 `src`가 `(src_len, batch_size)` 형태라고 주석으로 명시되어 있습니다.
        *   `src_len`: 원본 문장의 토큰(단어) 개수, 즉 문장의 길이입니다.
        *   `batch_size`: 한 번에 처리되는 문장의 개수입니다.
    *   이 `src`는 `encoder_tokenizer`에 의해 이미 숫자의 시퀀스(토큰 ID)로 변환된 상태입니다. 즉, 각 숫자는 어휘 사전(vocabulary) 내의 특정 단어를 가리킵니다.
    *   `forward` 메서드 내부에서는 이 `src` (토큰 ID 시퀀스)가 `self.embedding` 레이어를 통과하여 각 토큰 ID에 해당하는 밀집 임베딩 벡터(`embedded`)로 변환된 후, LSTM 레이어의 입력으로 사용됩니다.

### 디코더

````python
class Decoder(nn.Module):
    def __init__(self, output_dim, emb_dim, hidden_dim, attention, num_layers):
        super(Decoder, self).__init__()

        self.output_dim = output_dim
        self.attention = attention
        self.embedding = nn.Embedding(output_dim, emb_dim)
        self.rnn =  nn.LSTM(emb_dim + hidden_dim, hidden_dim, num_layers, batch_first=True)
        self.fc_out = nn.Linear(hidden_dim + hidden_dim, output_dim)

    def forward(self, input, hidden, cell, encoder_outputs):
        # input : (batch_size,)  # 현재 예측해야 할 단어의 바로 이전 단어 ID (배치 내 각 문장에 대한 하나의 단어 ID)
        # hidden : (batch_size, hidden_dim)
        # encoder_outputs : (batch_size, src_len, hidden_dim)
        input = input.unsqueeze(0)  # input : (1, batch_size)
        embedded = self.embedding(input)  # embedded : (1, batch_size, emb_dim)

        # attention distribution을 계산합니다.
        a = self.attention(hidden[-1], encoder_outputs)  # a : (batch_size, src_len), softmax의 출력
                           # hidden[-1]은 디코더의 LSTM 네트워크에서 가장 마지막 레이어의 은닉 상태
                           # 어텐션 메커니즘이 encoder_outputs에서 가장 관련성 높은 정보를 찾아내는 역할
                           # 이는 각 배치 내에서 encoder_outputs의 각 토큰에 대한 어텐션 가중치(Attention Weights)

        # H에 가중치를 부여해 attention value(Context vector) 계산
        # 현재 디코딩 시점에서 입력 시퀀스 전체에서 가장 관련성 높은 정보를 압축한 요약 벡터 역할
        a = a.unsqueeze(1)  # a : (batch_size, 1, src_len)
        context = torch.bmm(a, encoder_outputs)  # context : (batch_size, 1, hidden_dim), bmm: Batch Matrix Multiplication, 가중 평균 계산
        context = context.permute(1, 0, 2)  # context : (1, batch_size, hidden_dim)

        # embedded와 context를 결합하여 RNN에 전달
        rnn_input = torch.cat((embedded, context), dim=2) # (1, batch_size, emb_dim + hidden_dim)
        # rnn_input을 (batch_size, 1, emb_dim + hidden_dim) 형태로 permute합니다. (batch_first=True를 위해)
        rnn_input = rnn_input.permute(1, 0, 2)

        output, (hidden, cell) = self.rnn(rnn_input, (hidden, cell))

        # 출력층에서는 현재 hidden state와 context vector를 결합하여 예측값 생성
        output = output.squeeze(1)  # output : (batch_size, hidden_dim)
        context = context.squeeze(0)  # context : (batch_size, hidden_dim)
        prediction = self.fc_out(torch.cat((output, context), dim=1))  # (batch_size, output_dim)

        return prediction, hidden, cell, a.squeeze(1)
````

### 디코더 설명

#### 1. Decoder `forward` 메서드의 `input`과 `input.unsqueeze(0)` 설명

`Decoder`의 `forward` 메서드에서 `input`은 현재 시점에서 디코더가 예측할 다음 단어(토큰)를 위한 **이전 시점의 출력 단어(토큰) ID**를 나타냅니다.

`def forward(self, input, hidden, cell, encoder_outputs):`

**1. `input`은 무엇인가?**

*   **학습 시 (Training Time):** `input`은 보통 **정답 시퀀스(target sequence)**에서 현재 예측해야 할 단어의 바로 이전 단어 ID가 들어옵니다. 예를 들어, "나는 학생이다."를 번역할 때 디코더가 "I"를 예측하기 위해 `<bos>` 토큰을 입력으로 받고, "am"을 예측하기 위해 "I"를 입력으로 받는 식입니다. 이는 Teacher Forcing 전략에서 주로 사용됩니다.
*   **추론 시 (Inference Time):** `input`은 디코딩을 시작할 때 **시작 토큰 (BOS, Begin-of-Sequence)** ID로 초기화되고, 이후 각 시점에서는 디코더가 **직접 예측한 단어의 ID**가 다음 시점의 `input`으로 사용됩니다. 즉, 이전 스텝에서 모델이 가장 확률이 높다고 예측한 단어가 다음 스텝의 입력으로 재활용됩니다.

따라서 `input`은 `batch_size`만큼의 단어 ID를 담고 있는 텐서로, 그 형태는 `(batch_size,)` 입니다. 즉, 배치 내 각 문장에 대한 하나의 단어 ID를 가지고 있습니다.

**2. 왜 `input.unsqueeze(0)`를 하는가?**

`input.unsqueeze(0)`는 `input` 텐서의 가장 앞(0번째 인덱스)에 차원(dimension)을 하나 추가하는 역할을 합니다.

*   원래 `input`의 형태는 `(batch_size,)` 입니다.
*   `self.embedding` 레이어는 일반적으로 `(sequence_length, batch_size)` 또는 `(batch_size, sequence_length)` 형태의 입력을 기대합니다. `nn.Embedding`은 입력 텐서의 마지막 차원을 제외한 모든 차원을 시퀀스 길이로 간주할 수 있습니다.
*   여기서 `self.embedding = nn.Embedding(output_dim, emb_dim)`은 각 토큰 ID를 `emb_dim` 크기의 벡터로 변환합니다. 이를 위해 임베딩 레이어는 입력으로 받는 토큰 ID가 `(sequence_length, batch_size)` 또는 `(batch_size, sequence_length)` 형태로 들어올 것으로 예상합니다.
*   `input.unsqueeze(0)`를 호출하면 `(batch_size,)` 형태의 `input`이 `(1, batch_size)` 형태로 변경됩니다. 여기서 `1`은 현재 시점에서 처리하는 단어의 시퀀스 길이가 1임을 의미합니다 (즉, 한 번에 한 단어씩 디코딩). 이렇게 차원이 추가되어 `self.embedding` 레이어가 올바르게 임베딩을 수행할 수 있도록 입력 형태를 맞춰주는 것입니다.
*   이후 `embedded` 텐서는 `(1, batch_size, emb_dim)` 형태가 되어 `self.rnn` (LSTM)의 입력 형태에 맞게 됩니다 (만약 `batch_first=True`라면 `(batch_size, 1, emb_dim)`). 현재 코드에서는 `embedded`를 바로 `torch.cat`을 통해 `context`와 결합하고, 이후 `permute`를 통해 `(batch_size, 1, emb_dim + hidden_dim)` 형태로 만들어 `self.rnn`에 전달합니다.

#### 2. `self.attention(hidden[-1], encoder_outputs)` 에서 `hidden`과 `[-1]` 설명

`Decoder`의 `forward` 메서드 내에서 `self.attention(hidden[-1], encoder_outputs)` 호출 시 사용되는 `hidden`과 `[-1]`의 의미는 다음과 같습니다.

1.  **`hidden`은 무엇인가?**
    *   `hidden`은 `Decoder`의 `nn.LSTM`에서 출력되는 **모든 레이어의 은닉 상태(hidden states) 텐서**입니다. LSTM이 여러 층으로 쌓여 있을 때(`num_layers > 1`), `(hidden, cell)` 튜플 중 `hidden`은 `(num_layers, batch_size, hidden_dim)` 형태를 가집니다. 여기서 각 `num_layers` 차원은 해당 층의 은닉 상태를 나타냅니다.
    *   이 은닉 상태들은 디코더가 지금까지 생성한 시퀀스의 정보를 압축적으로 담고 있습니다. 어텐션 메커니즘은 이 `hidden` 정보를 사용하여 원본 문장(`encoder_outputs`)의 어느 부분에 집중할지 결정합니다.

2.  **`[-1]`을 한 이유는 무엇인가?**
    *   `hidden[-1]`은 여러 LSTM 레이어 중 **가장 마지막(최상위) 레이어의 은닉 상태**를 선택하는 것을 의미합니다.
    *   일반적으로 다층(multi-layer) LSTM에서는 가장 마지막 레이어의 은닉 상태가 이전 모든 레이어와 현재까지의 시퀀스 정보가 가장 잘 종합되어 있다고 간주됩니다. 따라서 이 최상위 레이어의 은닉 상태를 어텐션 메커니즘의 쿼리(query)로 사용하여 인코더 출력과 비교하고 어텐션 가중치를 계산하는 것이 일반적인 관행입니다.
    *   만약 `hidden`이 `(num_layers, batch_size, hidden_dim)` 형태라면, `hidden[-1]`은 `(batch_size, hidden_dim)` 형태의 텐서가 됩니다. 이 형태는 `BahdanauAttention` 클래스의 `forward` 메서드에서 `hidden` 인수로 예상하는 형태(`(batch_size, hidden_dim)`)와 일치합니다.

요약하자면, `hidden[-1]`은 디코더의 LSTM 네트워크에서 가장 마지막 레이어의 은닉 상태를 가져와, 이를 기반으로 어텐션 메커니즘이 인코더 출력(`encoder_outputs`)에서 가장 관련성 높은 정보를 찾아내도록 하는 역할을 합니다.

#### 3. `torch.bmm(a, encoder_outputs)`를 한 이유

`torch.bmm`은 **Batch Matrix Multiplication (배치 행렬 곱셈)**을 수행하는 함수입니다. `Decoder`의 `forward` 메서드에서 이 연산이 사용되는 주요 목적은 **컨텍스트 벡터(Context Vector)**를 계산하기 위함입니다.

`a`와 `encoder_outputs`의 형태를 다시 살펴보겠습니다:
*   `a`: `(batch_size, 1, src_len)` - 이는 각 배치 내에서 원본 시퀀스의 각 토큰에 대한 **어텐션 가중치(Attention Weights)**를 나타냅니다. `1`은 디코더의 현재 시점 하나를 의미합니다.
*   `encoder_outputs`: `(batch_size, src_len, hidden_dim)` - 이는 인코더의 모든 은닉 상태 출력으로, 원본 시퀀스의 각 토큰에 해당하는 정보를 포함합니다.

**`torch.bmm(a, encoder_outputs)` 연산의 역할:**

이 연산은 다음과 같은 의미를 가집니다:

1.  **가중 평균 계산:** `a`에 있는 어텐션 가중치를 `encoder_outputs`의 각 원본 토큰 벡터에 곱하여 가중 평균을 계산합니다.
    *   즉, `decoder`가 현재 시점에서 다음 단어를 예측하기 위해 `encoder_outputs` 중 어떤 부분에 얼마나 집중해야 하는지(가중치 `a`)를 반영하여, 이 '집중할 부분'들을 모두 합산하는 것입니다.
2.  **컨텍스트 벡터 생성:** 이 가중 평균의 결과가 바로 **컨텍스트 벡터(Context Vector)**입니다. 컨텍스트 벡터는 현재 디코딩 시점에서 원본 시퀀스 전체에서 가장 관련성 높은 정보를 압축한 요약 벡터 역할을 합니다.
    *   연산 후 `context`의 형태는 `(batch_size, 1, hidden_dim)`이 됩니다. 이는 각 배치별로 디코더의 현재 예측을 위한 단일 컨텍스트 벡터를 나타냅니다.

**결론적으로,** `torch.bmm(a, encoder_outputs)`는 어텐션 메커니즘의 핵심 단계로, 디코더가 원본 문장의 정보를 효과적으로 활용하여 더 정확한 번역을 생성할 수 있도록 현재 디코딩 시점에 가장 적합한 '문맥 정보'를 요약하여 추출하는 역할을 합니다.

#### 4. `output, (hidden, cell) = self.rnn(rnn_input, (hidden, cell))` 에서 `output` 의미

`Decoder`의 `forward` 메서드에서 `output, (hidden, cell) = self.rnn(rnn_input, (hidden, cell))` 라인에서 `output`은 `self.rnn` (LSTM) 레이어의 **현재 시점(time step)에 대한 출력 은닉 상태(output hidden state)**를 의미합니다.

각 요소의 역할은 다음과 같습니다:

*   **`rnn_input`**: 현재 디코딩 시점의 입력으로, 이전 시점의 임베딩된 단어(`embedded`)와 어텐션 메커니즘을 통해 계산된 컨텍스트 벡터(`context`)가 결합된 형태입니다.
*   **`(hidden, cell)` (입력)**: 이전 시점의 LSTM 은닉 상태(`hidden`)와 셀 상태(`cell`)입니다. 이들은 LSTM이 과거의 정보를 기억하는 데 사용됩니다.

*   **`output` (출력)**: `self.rnn`은 `rnn_input`을 처리하고 이전 상태 `(hidden, cell)`를 업데이트한 후, 현재 시점의 출력 은닉 상태를 반환합니다. `batch_first=True`로 설정된 LSTM의 경우, `output`의 형태는 `(batch_size, sequence_length, hidden_dim)`이 됩니다. 디코더에서는 한 번에 한 단어씩 예측하므로 `sequence_length`가 1이 되어, `(batch_size, 1, hidden_dim)` 형태를 가집니다.

    이 `output`은 현재 시점의 입력과 이전 상태를 바탕으로 LSTM이 계산한 '현재 상태에 대한 요약 정보'로 볼 수 있습니다. 이 정보는 이후 최종 출력층(`self.fc_out`)으로 전달되어 실제 다음 단어 예측을 위한 중요한 피처로 사용됩니다.

*   **`(hidden, cell)` (출력)**: `self.rnn`이 반환하는 `(hidden, cell)`은 다음 시점의 디코딩에 사용될 업데이트된 은닉 상태와 셀 상태입니다. 이들은 디코더가 시퀀스를 계속 생성해나가면서 문맥을 유지하는 데 필요합니다.

요약하자면, `output`은 LSTM이 현재 시점에서 생성한 내부 표현으로, 컨텍스트 벡터와 함께 최종적으로 다음에 올 단어를 예측하는 데 사용되는 핵심적인 정보입니다.

#### 5. `prediction = self.fc_out(torch.cat((output, context), dim=1))` 의미

이 코드는 디코더가 현재 시점에서 다음 단어를 예측하기 위한 최종 단계입니다.

각 부분을 살펴보겠습니다:

1.  **`output`**: 이전 질문에서 설명했듯이, `output`은 `self.rnn` (LSTM)의 현재 시점 출력 은닉 상태(output hidden state)입니다. `(batch_size, hidden_dim)` 형태를 가집니다. 이 `output`은 디코더의 LSTM이 현재까지의 디코딩 시퀀스와 입력 `rnn_input` (이전 단어 임베딩 + 컨텍스트 벡터)를 처리하여 얻은 내부 표현입니다.

2.  **`context`**: 어텐션 메커니즘을 통해 계산된 컨텍스트 벡터입니다. `(batch_size, hidden_dim)` 형태를 가집니다. 이 `context`는 인코더의 모든 출력 중에서 현재 디코딩 시점에서 가장 중요하다고 판단된 정보들을 압축한 요약 벡터입니다.

3.  **`torch.cat((output, context), dim=1)`**:
    *   `torch.cat`은 두 개 이상의 텐서를 특정 차원(dimension)을 따라 이어붙이는(concatenate) 함수입니다.
    *   `output`과 `context`는 모두 `(batch_size, hidden_dim)` 형태입니다.
    *   `dim=1`은 `hidden_dim` 차원을 따라 이어붙이겠다는 의미입니다.
    *   결과적으로 `(batch_size, hidden_dim + hidden_dim)` 즉, `(batch_size, 2 * hidden_dim)` 형태의 새로운 텐서가 생성됩니다. 이 텐서는 LSTM의 현재 상태 정보와 어텐션이 강조한 원본 문맥 정보를 모두 포함하게 됩니다.

4.  **`self.fc_out(...)`**:
    *   `self.fc_out`은 `nn.Linear(hidden_dim + hidden_dim, output_dim)`으로 정의된 **선형(Fully Connected) 레이어**입니다.
    *   이 레이어의 역할은 `torch.cat`으로 결합된 `(batch_size, 2 * hidden_dim)` 형태의 입력 벡터를 받아, 디코더의 출력 어휘 사전 크기(`output_dim`)에 해당하는 `(batch_size, output_dim)` 형태의 벡터로 변환하는 것입니다.
    *   이 `output_dim` 크기의 벡터는 각 단어가 다음에 올 확률을 나타내는 로짓(logits)입니다. 즉, 이 벡터의 각 요소는 특정 단어가 다음에 올 확률에 비례하는 값을 가집니다.

**요약하자면,**
디코더는 현재까지의 디코딩 정보를 담고 있는 LSTM의 `output`과, 어텐션 메커니즘이 원본 문장에서 중요하다고 판단한 `context` 벡터를 **결합**합니다. 이 결합된 정보를 선형 레이어(`self.fc_out`)에 통과시켜 **다음에 올 단어에 대한 확률 분포(또는 로짓)**를 계산합니다. 이 확률 분포에서 가장 높은 확률을 가진 단어가 현재 시점에서 디코더가 예측하는 단어가 됩니다.

### Seq2SeqAttention model 설계

````python
class Seq2SeqAttention(nn.Module):
    def __init__(self, encoder, decoder, device):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, src, trg=None, max_len=30, bos_id = 1, eos_id=2):
        # 학습 모드에서는 trg_len 사용, 추론 모드에서는 max_len까지 동적 생성
        batch_size = src.shape[1] # src가 (src_len, batch_size) 형태라고 가정
        trg_vocab_size = self.decoder.fc_out.out_features

        # 조기 종료를 위해 tensor가 아닌 리스트 사용
        outputs = []

        # 시각화를 위해 attention 저장
        attentions = []

        # 인코더를 통해 context 생성
        encoder_outputs, hidden, cell = self.encoder(src) # h_0와 c_0를 hidden, cell로 받음

        if trg is not None:
            for t in range(0, trg.shape[0]):
                input = trg[t]
                output, hidden, cell, attention = self.decoder(input, hidden, cell, encoder_outputs) # hidden과 cell을 모두 전달
                outputs.append(output.unsqueeze(0))
                attentions.append(attention.unsqueeze(0))

        else:
		    # inference에서는 target(정답)이 없기 때문에 sos_token을 생성해줍니다.
            input = torch.full((batch_size,), bos_id, dtype=torch.long, device=self.device)
            finished = torch.zeros(batch_size, dtype=torch.bool, device=self.device)

            for t in range(max_len):
                output, hidden, cell, attention = self.decoder(input, hidden, cell, encoder_outputs) # hidden과 cell을 모두 전달
                outputs.append(output.unsqueeze(0))
                attentions.append(attention.unsqueeze(0))
                top1 = output.argmax(1)
                input = top1

                # 조기 종료 조건
                finished |= (top1 == eos_id)
                if finished.all():
                    break

        outputs = torch.cat(outputs, dim=0)  # (trg_len, batch_size, output_dim)
        attentions = torch.cat(attentions, dim=0)  # (trg_len, batch_size, src_len)

        return outputs, attentions
````

````python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

input_dim = len(encoder_tokenizer)
output_dim = len(decoder_tokenizer)
emb_dim = 256   # 256
hid_dim = 512  # 512
NUM_LAYERS = 3  # LSTM Layers = 3
````

````python
encoder = Encoder(input_dim, emb_dim, hid_dim, NUM_LAYERS).to(device)
attention = BahdanauAttention(hid_dim).to(device)
decoder = Decoder(output_dim, emb_dim, hid_dim, attention, NUM_LAYERS).to(device)
model = Seq2SeqAttention(encoder, decoder, device).to(device)
````

## 4. 훈련하기
### (1) Optimizer & Loss

````python
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss(ignore_index=pad_id)

print("슝~")
````

### (2) train_step 구현하기

````python
def train_step(model, data_loader, optimizer, criterion, epoch):
    model.train()
    epoch_loss = 0

    progress_bar = tqdm(data_loader, desc=f"Epoch {epoch+1}", leave=True)

    for src, trg_input, trg_label in progress_bar:
        # 모델의 입력 순서에 맞게 transpose 변환
        src = src.permute(1, 0).to(device)
        trg_input = trg_input.permute(1, 0).to(device)
        trg_label = trg_label.permute(1, 0).to(device)
        optimizer.zero_grad()

        outputs,_ = model(src, trg_input)

        # (trg_len, batch_size, output_dim)을 (batch_size * trg_len, output_dim)으로 변환
        outputs = outputs.reshape(-1, outputs.shape[-1])
        trg_label = trg_label.reshape(-1)

        loss = criterion(outputs, trg_label)
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1)

        optimizer.step()

        epoch_loss += loss.item()

        progress_bar.set_postfix(loss=loss.item())

    return epoch_loss / len(data_loader)

print("슝~")
````

### (3) 훈련 시작하기

````python
def eval_step(model, data_loader, optimizer, criterion):
    model.eval()
    total_loss = 0

    for src, trg_input, trg_label in data_loader:
        # 모델의 입력 순서에 맞게 transpose 변환
        src = src.permute(1, 0).to(device)
        trg_input = trg_input.permute(1, 0).to(device)
        trg_label = trg_label.permute(1, 0).to(device)

        outputs,_ = model(src, trg_input)

        # (trg_len, batch_size, output_dim)을 (batch_size * trg_len, output_dim)으로 변환
        outputs = outputs.reshape(-1, outputs.shape[-1])
        trg_label = trg_label.reshape(-1)

        loss = criterion(outputs, trg_label)

        total_loss += loss.item()

    return total_loss / len(data_loader)

print("슝~")
````

````python
%%time

EPOCHS = 15  # 20 > 10 > 15

for epoch in range(EPOCHS):
    train_loss = train_step(model, train_loader, optimizer, criterion, epoch)
    if validation_loader:
        valid_loss = eval_step(model, validation_loader, optimizer, criterion)
        print(f'Epoch {epoch+1}/{EPOCHS}, Train Loss: {train_loss:.4f}, Validation Loss: {valid_loss:.4f}')
    else:
        print(f'Epoch {epoch+1}/{EPOCHS}, Train Loss: {train_loss:.4f}')
````

### 번역 성능 평가 + Attention Map 시각화

````python
def evaluate(sentence, model, encoder_tokenizer, decoder_tokenizer, max_len=30):
    model.eval()

    sentence = preprocess_sentence(sentence)
    src_ids = encoder_tokenizer.encode(sentence)
    src_ids = src_ids[:max_len]
    src_ids = src_ids + [0] * (max_len - len(src_ids))  # 패딩 추가
    src_tensor = torch.tensor(src_ids).unsqueeze(1).to(device)  # (src_len, 1)

    with torch.no_grad():
        outputs, attentions = model(src_tensor, max_len=max_len)

    result = [decoder_tokenizer.decode([token.item()]) for token in outputs.argmax(2).squeeze(1)]

    if "<end>" in result:
        result = result[:result.index("<end>")]

    return result, sentence, attentions.squeeze(1).cpu().numpy()
````

````python
def plot_attention(attention, sentence, predicted_sentence):
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.matshow(attention, cmap='viridis')

    fontdict = {'fontsize': 14}

    ax.set_xticks(range(len(sentence)))
    ax.set_xticklabels(sentence, fontdict=fontdict, rotation=90)

    ax.set_yticks(range(len(predicted_sentence)))
    ax.set_yticklabels(predicted_sentence, fontdict=fontdict)

    plt.show()
````

````python
def translate(sentence, model, encoder_tokenizer, decoder_tokenizer, max_len=30):
    result, sentence, attention = evaluate(sentence, model, encoder_tokenizer, decoder_tokenizer, max_len)

    print('Input: %s' % (sentence))
    print('Predicted translation: {}'.format(result))

    # Attention 크기 조정 (trg_len, src_len)
    attention = attention[:len(result), :len(sentence.split())]

    plot_attention(attention, sentence.split(), result)
````

````python
translate("오바마는 대통령이다.", model, encoder_tokenizer, decoder_tokenizer, max_len=30)
````

````python
translate("시민들은 도시 속에 산다.", model, encoder_tokenizer, decoder_tokenizer, max_len=30)
````
